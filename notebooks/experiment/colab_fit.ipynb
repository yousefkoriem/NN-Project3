```python
import pickle
import pandas as pd
import numpy as np
import joblib
import keras
import tensorflow as tf
```

```python
(x_train,y_train) = pickle.load(open('data/processed/train.pkl','rb'))
(x_test,y_test) = pickle.load(open('data/processed/test.pkl','rb'))
```

```python
from keras import callbacks, layers, models, regularizers, optimizers, callbacks,losses,metrics
```

```python
checkpoint = callbacks.ModelCheckpoint(
    filepath='models/architecture/best_cifar10_model.keras',
    monitor='val_accuracy',
    save_best_only=True,
    verbose=1
)
```

```python
early_stopping = callbacks.EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)
```

```python
reduce_lr = callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6,
    verbose=1
)
```

```python
model = models.load_model('models/architecture/untrained_cifar10_model.keras')
```

```python
history = model.fit(
    x=x_train,
    y=y_train,
    batch_size=128,
    epochs=100,
    validation_split=0.2,
    callbacks=[checkpoint, early_stopping, reduce_lr]
)
```

```python
history_df = pd.DataFrame(history.history)
history_df.to_csv('results/tables/training_history.csv', index=False)
```

```python
score = model.evaluate(x_test, y_test)
score_df = pd.DataFrame({'loss': [score[0]], 'accuracy': [score[1]],'f1_score': [score[2]]})
score_df.to_csv('results/tables/test_score.csv', index=False)
```

```python
model.save('models/architecture/final_cifar10_model.keras')
```
