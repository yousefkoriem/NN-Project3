```python
import pickle
import keras
import numpy as np
from keras import layers, models, regularizers, optimizers, callbacks,losses,metrics
```

```python
input1 = layers.Input(shape=(32,32,3))
```

```python
aug = layers.RandomFlip("horizontal")(input1)
aug = layers.RandomRotation(0.1)(aug)
aug = layers.RandomZoom(0.1)(aug)
aug = layers.RandomTranslation(height_factor=0.1, width_factor=0.1)(aug)
```

```python
cnn_params = {
    'kernel_size': (3,3),
    'activation': 'relu',
    'use_bias': False
}
```

```python
cnn = layers.Conv2D(filters=32,padding='same',**cnn_params)(aug)
cnn = layers.BatchNormalization()(cnn)

cnn = layers.Conv2D(filters=32,**cnn_params)(cnn)
cnn = layers.BatchNormalization()(cnn)

cnn = layers.MaxPooling2D()(cnn)
cnn = layers.Dropout(0.2)(cnn)
```

```python
cnn = layers.Conv2D(filters=64,padding='same',**cnn_params)(cnn)
cnn = layers.BatchNormalization()(cnn)

cnn = layers.Conv2D(filters=64,**cnn_params)(cnn)
cnn = layers.BatchNormalization()(cnn)

cnn = layers.MaxPooling2D()(cnn)
cnn = layers.Dropout(0.25)(cnn)
```

```python
cnn = layers.Conv2D(filters=128,padding='same',**cnn_params)(cnn)
cnn = layers.BatchNormalization()(cnn)

cnn = layers.Conv2D(filters=128,**cnn_params)(cnn)
cnn = layers.BatchNormalization()(cnn)

cnn = layers.MaxPooling2D()(cnn)
cnn = layers.Dropout(0.3)(cnn)
```

```python
ann_params = {
    'activation': 'relu',
    'use_bias': False
}
```

```python
ann = layers.Flatten()(cnn)
```

```python
ann = layers.Dense(units=256,**ann_params)(ann)
ann = layers.BatchNormalization()(ann)
ann = layers.Dropout(0.5)(ann)

ann = layers.Dense(units=128,**ann_params)(ann)
ann = layers.BatchNormalization()(ann)
ann = layers.Dropout(0.3)(ann)

ann = layers.Dense(units=32,**ann_params)(ann)
ann = layers.BatchNormalization()(ann)
```

```python
output1 = layers.Dense(units=10,activation='softmax')(ann)
```

```python
model = models.Model(inputs=input1,outputs=output1)
```

```python
model.compile(
    optimizer=optimizers.AdamW(learning_rate=1e-3,weight_decay=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy',metrics.F1Score(average='weighted')]
)
```

```python
model(np.zeros((1,32,32,3)))
```
